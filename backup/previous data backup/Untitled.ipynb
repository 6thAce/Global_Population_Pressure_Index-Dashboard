```python
import pandas as pd
```

```python
pop = pd.read_csv("fact_population_un.csv")
econ = pd.read_csv("fact_economics_imf.xls.csv")
water = pd.read_csv("fact_water_access.csv")
base = pd.read_csv("fact_country_year.csv")
country = pd.read_csv("dim_country_lookup.csv")
```

```python
pop = pop.rename(columns={
    "Location": "country",
    "Time": "year",
    "TPopulation1Jan": "population",
    "PopGrowthRate": "population_growth_rate"
})

pop = pop[["country", "year", "population", "population_growth_rate"]]
```

```python
pop["year"] = pd.to_numeric(pop["year"], errors="coerce")
pop["population"] = pd.to_numeric(pop["population"], errors="coerce")
pop["population_growth_rate"] = pd.to_numeric(pop["population_growth_rate"], errors="coerce")
```

```python
pop = pop.dropna()
pop = pop.drop_duplicates(subset=["country", "year"])
```

```python
water = water.rename(columns={
    "name": "country",
    "wat_basal_t": "basic_water_access_total"
})

water = water[["country", "year", "basic_water_access_total"]]

water["year"] = pd.to_numeric(water["year"], errors="coerce")
water["basic_water_access_total"] = pd.to_numeric(water["basic_water_access_total"], errors="coerce")

water = water.dropna()
```

```python
base = base[["country", "year", "water_consumption_per_capita"]]

base["year"] = pd.to_numeric(base["year"], errors="coerce")
base["water_consumption_per_capita"] = pd.to_numeric(base["water_consumption_per_capita"], errors="coerce")

base = base.dropna()
```

```python
base.columns
```

```python
base = base.rename(columns={"COUNTRY": "country"})
```

```python
base = base.melt(
    id_vars=["country"],
    var_name="year",
    value_name="water_consumption_per_capita"
)
```

```python
base["year"] = pd.to_numeric(base["year"], errors="coerce")
base["water_consumption_per_capita"] = pd.to_numeric(base["water_consumption_per_capita"], errors="coerce")
```

```python
base = base.dropna()
base = base.drop_duplicates(subset=["country", "year"])
```
